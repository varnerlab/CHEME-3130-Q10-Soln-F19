#### Estimate the saturation pressure and $\Delta{H}_{i}^{vap}$ for Cornellium at T = 310K 

Integrating: 

\begin{equation}
\frac{dP_{i}^{sat}}{P_{i}^{sat}} = \left(\frac{\Delta{H}_{vap,i}}{R}\right)\frac{dT}{T^{2}}
\end{equation} gives:

\begin{equation}
\ln P_{i}^{sat} = \mathcal{C} - \frac{\Delta{H}_{i}^{vap}}{R}\left(\frac{1}{T}\right)
\end{equation}where we have assumed the heat of vaporization is not a strong function of temperature (assumption 3 of the Clausius-Clapeyron derivation. 

The indefinite form of the Clausius-Clapeyron equation is a line in the 1/T -vs- $\ln P_{i}^{sat}$ plane. We can plot the data given in the problem to calculate the slope (heat of vaporization), and then project back to the axis to get the integration constant $\mathcal{C}$
